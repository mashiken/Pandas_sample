## データ集計

```python
import pandas as pd
```

```python
pd.options.display.max_rows = 10
pd.options.display.max_columns = None
```

```python
df = pd.read_excel('sample.xlsx',sheet_name='実績管理表_売上欠損')
df
```

```python
df.groupby('氏名')
```

```python
df.groupby('氏名').mean()
```

```python
df[['氏名','売上金額']].groupby('氏名').mean()
# 氏名 ごとに「売上金額」の平均を算出
```

```python
pd.options.display.float_format = '{:.0f}'.format
# 小数点以下の数字を表示させない
```

```python
df[['氏名','売上金額']].groupby('氏名').mean() # 平均
```

```python
df[['氏名','売上金額']].groupby('氏名').sum() # 合計
```

```python
df[['氏名','売上金額']].groupby('氏名').count() # データ数
# count メソッドでは欠損値は数えない
```

```python
df[['氏名','売上金額']].groupby('氏名').size() # データ数 欠損値あり
```

```python
df[['氏名','売上金額']].groupby('氏名').nth(5)
```

```python
df[['氏名','売上金額']].groupby('氏名').nth(0)
```

```python
df[['氏名','売上金額']].groupby('氏名').max() # 最大値
```

```python
df[['氏名','売上金額']].groupby('氏名').min() # 最小値
```

```python
## 中央値 median 標準偏差 std 分散 var を求める事が可能
```

```python
df[['氏名','商品分類','売上金額']].groupby(['氏名','商品分類']).mean()
```

```python
df[['氏名','商品分類','売上金額']].groupby(['氏名','商品分類'],as_index=False).mean()
```

```python
df[['氏名','売上金額']].groupby(['氏名']).mean()
```

```python
df[['氏名','売上金額']].groupby(['氏名']).agg('mean')
# agg メソッド 平均・合計を1度に複数の処理を適用できる
```

```python
df[['氏名','売上金額']].groupby(['氏名']).agg(['mean','sum'])
```

```python
df_group = df[['氏名','売上金額']].groupby(['氏名']).agg(['mean','sum'])
df_group.applymap('{:,.0f}'.format) #3桁区切り
```

```python
df_group = df[['氏名','売上金額']].groupby(['氏名']).agg(['mean','sum','count','max','min','std','var'])
df_group.applymap('{:,.0f}'.format) #3桁区切り
```

```python
df_group = df[['氏名','売上金額']].groupby(['氏名']).agg(['describe'])
df_group.applymap('{:,.0f}'.format) #3桁区切り
```

```python
import numpy as np # 独自の関数を指定できる その中で税率を計算したり、自社風にアレンジする事が可能

def cal_tax(s):
    return np.sum(s)*1.10
```

```python
df_group = df[['氏名','売上金額']].groupby(['氏名']).agg({'売上金額':cal_tax})
df_group.applymap('{:,.0f}'.format) #3桁区切り
```
