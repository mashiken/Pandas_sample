```python
#concat DataFrameを縦に結合
```

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
# warningを表示しない設定をしている
```

```python
df01 = pd.DataFrame( {'氏名':['佐藤','鈴木','高橋','田中'],
                      'クラス':['df01','df01','df01','df01'],
                      '数学':[1,2,3,4],
                      '国語':[5,6,7,8]})
df01
```

```python
df02 = pd.DataFrame( {'氏名':['佐藤','鈴木','山本'],
                      'クラス':['df02','df02','df02'],
                      '数学':[9,10,11],
                      '国語':[12,13,14]})
df02
```

```python
pd.concat([df01,df02])
```

```python
# index番号を再度振りなおす
pd.concat([df01,df02], ignore_index=True)
```

```python
df03 = pd.DataFrame( {'氏名':['佐藤','鈴木','加藤'],
                      'クラス':['df03','df03','df03'],
                      '数学':[15,16,17],
                      '国語':[18,19,20]})
df03
```

```python
# 3つのデータフレーム結合
pd.concat([df01,df02,df03])
```

```python
df = pd.concat([df01,df02,df03],keys=['1番目','2番目','3番目'])
df
```

```python
df.index
```

```python
# カラムが共通していない場合
df04 = pd.DataFrame( {'氏名':['佐藤','鈴木','佐々木'],
                      'クラス':['df04','df04','df04'],
                      '数学':[21,22,23],
                      '社会':[24,25,26]})
df04
```

```python
df01
```

```python
pd.concat([df01,df04])
```

```python
pd.concat([df01,df04], join='outer')
```

```python
pd.concat([df01,df04], join='inner')
```

```python
pd.concat([df01,df04], axis=0)
# axis0で縦方向、1で横方向
```

```python
pd.concat([df01,df04], axis='index')
# indexは縦方向に振られる、縦に結合
```

```python
pd.concat([df01,df04], axis='columns')
#columnsは横に振られる、横に結合
```

```python
pd.concat([df01,df02,df03,df04], axis='columns')
```

```python
pd.concat([df01,df02,df03,df04], axis='columns', join='inner')
# join にinnerを指定している
```

```python
# データフレームのインデックスを氏名に変更
df01 = df01.set_index('氏名')
df02 = df02.set_index('氏名')
df03 = df03.set_index('氏名')
df04 = df04.set_index('氏名')
```

```python
df04
```

```python
pd.concat([df01,df02,df03,df04], axis='columns', join='outer')
```
