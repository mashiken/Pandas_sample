```python
# データフレームワークの結合
```

```python
import pandas as pd
```

```python
df01 = pd.DataFrame( {'氏名':['高橋','伊藤','鈴木','佐藤'],
                      'クラス':['a001','a001','a001','a001'],
                      '数学':[1,2,3,4]})
df01
```

```python
df02 = pd.DataFrame( {'氏名':['高橋','伊藤','鈴木','佐藤'],
                      'クラス':['a001','a001','a001','a001'],
                      '国語':[5,6,7,8]})
df02
```

```python
pd.merge(df01,df02, on='氏名')
# 氏名は共通キーに設定されている。クラスは指定していない為、重複
```

```python
pd.merge(df01 ,df02, on='氏名', suffixes=['_left','_right'])
# suffixesで名前を付ける
```

```python
pd.merge(df01 ,df02, on='氏名', suffixes=['','重複'])
```

```python
df02 = pd.DataFrame( {'名前':['高橋','伊藤','鈴木','佐藤'],
                      'クラス':['a001','a001','a001','a001'],
                      '国語':[5,6,7,8]})
df02
```

```python
pd.merge(df01, df02, left_on='氏名', right_on='名前')
# カラム名が違うもの同士を結合している
```

```python
df03 = pd.DataFrame( {'名前':['高橋','高橋'],
                      'クラス':['a001','a002'],
                      '英語':[15,16]})
df03
```

```python
pd.merge(df01, df03, left_on=['氏名','クラス'], right_on=['名前','クラス'])
```

```python
#inner 2つのデータフレームの共通しているキーのみ結合
#right 右のデータフレームをベースに結合(右は全て残り、左は共通部分のみ)
#left 左のデータフレームをベースに結合(左は全て残り、右は共通部分のみ)
#outer 共通しているものだけ結合。しかししていない部分も残す
```

```python
df01
```

```python
df04 = pd.DataFrame( {'氏名':['高橋','伊藤','渡辺','加藤'],
                      'クラス':['a001','a001','a001','a001'],
                      '国語':[5,6,7,8]})
df04
```

```python
pd.merge(df01, df04, on='氏名',how='left')
```

```python
pd.merge(df01, df04, on='氏名',how='right')
```

```python
pd.merge(df01, df04, on='氏名',how='outer')
```

```python
pd.merge(df01, df04, on='氏名',how='inner')
# デフォルトではinner
```

```python
pd.merge(df01, df04, on='氏名', how='outer', indicator=True)
# indicator=True 元のデータフレームの情報を表示
```

```python
# 結合キーに関してはカラム名しか指定出来ない
```

```python
df05 = pd.DataFrame({'クラス':['a001','a001','a001','a001'],
                     '数学':[1,2,3,4]}
                   ,index = ['高橋','伊藤','鈴木','佐藤'])
df05
```

```python
df02
```

```python
pd.merge(df05, df02, left_index=True, right_on='名前')
```

```python
pd.merge(df05, df02, left_index=True, right_on='名前', sort=True)
```
