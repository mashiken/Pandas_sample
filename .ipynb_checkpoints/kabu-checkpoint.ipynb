```python
from pandas_datareader import data 
#installしたpandas_datareaderをdataとしてimportして使用できるようにする
import pandas as pd
#pandas データ解析を支援する機能を提供するpythonのライブラリ
import matplotlib.pyplot as plt
#matplotlib グラフを描くことができる matplotlibのpyplotをimport

%matplotlib inline
# jupyterlab上で作成したグラフを表示
# エラーが出た場合は以下をコメントアウト
# pd.core.common.is_list_like = pd.api.types.is_list_like
```

```python
# 日経平均のデータを取得
start = '2019-06-01'
end = '2020-06-01'

df = data.DataReader('^N225','yahoo', start, end)
#('取得したいティッカーシンボル','データソースの名前','開始日','終了日')
```

```python
df.head(10)
#headで上位10件
#heigh高値/Low安値/Open始値/Close終値/volume取引出来高/AdjClose調整後の終値
```

```python
price=df['Adj Close']
```

```python
# 線グラフに名前をつける
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
# 線グラフ1つ1つの名前を凡例という 今回は'Nikkei225'
plt.legend()
```

```python
# グラフにタイトル
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.title('N225',color='blue',backgroundcolor='green',size=40,loc='center')
# ('名前','文字色'，'背景','大きさ'，'位置')
plt.legend()
```

```python
# x軸 ｙ軸名前変更
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')
plt.title('N225',color='blue',backgroundcolor='green',size=40,loc='center')

plt.xlabel('date', color='black', size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
## グラフをいつ買って、いつ売るかの判断が難しい
# 単純移動平均の指標を追加する
# 移動平均は過去〇〇日分の平均
# 加重移動平均 6月7日なら6月7日に近い日付のデータ程価値が高い
# 指数平滑移動平均 同じく近い日付に重みをつける
```

```python
span01=5
span02=25
span03=50
# 移動平均の値(枠) 数字日間の単純移動平均を求める

#pythonで移動平均を算出するならrolingメソッドを利用する
df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()
#.min なら過去〇日間の最小値、.maxなら過去〇日間の最大値
```

```python
pd.set_option('display.max_rows',None)
# 行の省略をさせない
df.head(100)
```

```python
plt.figure(figsize=(30,10))
plt.plot(date,price,label='Nikkei225')

plt.plot(date,df['sma01'],label='sma01')
plt.plot(date,df['sma02'],label='sma02')
plt.plot(date,df['sma03'],label='sma03')

plt.title('N225',color='blue',backgroundcolor='green',size=40,loc='center')
plt.xlabel('date', color='black', size=30)
plt.ylabel('price', color='black', size=30)
plt.legend()
```

```python
# ゴールデンクロス
# 短期の移動平均線が中期以上の移動平均線を下から上に抜けていること(価格上昇のサイン)
# デッドクロス
# 短期の移動平均線が中期以上の移動平均線を上から下に抜けること(価格下降のサイン)
```

```python
# 棒グラフの作成方法
# 出来高を見るように設定する
plt.figure(figsize=(30,15))
plt.bar(date,df['Volume'],label='Volume',color='grey')

plt.legend()
```

```python
# 線グラフ・棒グラフを二分かつして表示
plt.figure(figsize=(30,15))
# グラフ領域を二分割 (縦方向分割(上・下で2),横方向分割，グラフ配置位置)
plt.subplot(2,1,1)
# 線グラフ表示
plt.plot(date,price,label='Close', color='#99b898')
plt.plot(date,df['sma01'],label='sma01', color='#e84a5f')
plt.plot(date,df['sma02'],label='sma02', color='#ff847c')
plt.plot(date,df['sma03'],label='sma03', color='#feceab')
plt.legend()
# 棒グラフ表示
plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend()
```

```python
# リクルートホールディングス 可視化
df = data.DataReader('6098.JP','stooq')

date=df.index
price=df['Close']

span01=5
span02=25
span03=50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()

plt.figure(figsize=(30,15))
plt.subplot(2,1,1)

plt.plot(date,price,label='Close',color='#99b898')
plt.plot(date,df['sma01'],label='sma01',color='#e84a5f')
plt.plot(date,df['sma02'],label='sma02',color='#ff847c')
plt.plot(date,df['sma03'],label='sma03',color='#feceab')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend()
```

```python
start = '2019-06-01'
end = '2020-05-01'
conpany_code = '6502.JP'
```

```python
df = data.DataReader(conpany_code,'stooq')
df = df[(df.index>=start) & (df.index<=end)]


date=df.index
price=df['Close']

span01=5
span02=25
span03=50

df['sma01'] = price.rolling(window=span01).mean()
df['sma02'] = price.rolling(window=span02).mean()
df['sma03'] = price.rolling(window=span03).mean()

plt.figure(figsize=(30,15))
plt.subplot(2,1,1)

plt.plot(date,price,label='Close',color='#99b898')
plt.plot(date,df['sma01'],label='sma01',color='#e84a5f')
plt.plot(date,df['sma02'],label='sma02',color='#ff847c')
plt.plot(date,df['sma03'],label='sma03',color='#feceab')
plt.legend()

plt.subplot(2,1,2)
plt.bar(date,df['Volume'],label='Volume',color='grey')
plt.legend()
```

```python
# 関数にまとめる
def company_stock(start,end,company_code):
    df = data.DataReader(conpany_code,'stooq')
    df = df[(df.index>=start) & (df.index<=end)]


    date=df.index
    price=df['Close']

    span01=5
    span02=25
    span03=50

    df['sma01'] = price.rolling(window=span01).mean()
    df['sma02'] = price.rolling(window=span02).mean()
    df['sma03'] = price.rolling(window=span03).mean()

    plt.figure(figsize=(20,10))
    plt.subplot(2,1,1)

    plt.plot(date,price,label='Close',color='#99b898')
    plt.plot(date,df['sma01'],label='sma01',color='#e84a5f')
    plt.plot(date,df['sma02'],label='sma02',color='#ff847c')
    plt.plot(date,df['sma03'],label='sma03',color='#feceab')
    plt.legend()

    plt.subplot(2,1,2)
    plt.bar(date,df['Volume'],label='Volume',color='grey')
    plt.legend()
```
