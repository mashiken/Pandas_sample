### 時系列データの扱い方

```python
import pandas as pd
```

```python
# 表示する行数を10行、列数を無制限に設定
pd.options.display.max_rows = 10
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('sample.csv', encoding='utf-8', header=0)
df
```

```python
df.info()
```

```python
# 売上日カラムを時系列データへ変換する
df['売上日'] = pd.to_datetime(df['売上日'])
df['売上日'] #datetime64
```

```python
# 売上日カラムをindexに指定する
df_2 = df.set_index("売上日")
df_2
```

```python
df_2.info()
```

```python
df02 = pd.read_csv('sample02.csv', encoding='utf-8', header=0)
df02 # header=0 で1行目をカラムにして読み込んでいる
```

```python
df02.info()
```

```python
df02['売上日'] = pd.to_datetime(df02['売上日'])
# 結果はエラーになる
# 年月日のカラムでは変換できない為
```

```python
df02['売上日'] = pd.to_datetime(df02['売上日'], format='%Y年%m月%d日')
df02['売上日']
```

```python
# csvを読み込む段階で、時系列データを変換する事は可能
# parse_dates 読み込みと同時に日付を時系列データに変換する
# index_col 指定した値をindexにする
df = pd.read_csv('sample.csv', encoding='utf-8', header=0, parse_dates=True, index_col='売上日')
df
```

```python
df.info()
```

### 時系列データの抽出

```python
df['2020-08-15':'2020-09-14']
```

### 時系列データの集計

```python
# 売上金額のデータのみ抽出
df_3 = df[['売上金額']]
df_3
```

```python
# 期間事の集計はresampleメソッド ()の中に期間を入れる
df_3.resample('M')
```

```python
df_3.resample('M').sum()
```

```python
df_3.resample('Q').sum()
```

```python
df_3.resample('10D').sum()
```

```python
df_3.resample('M').mean()
```

```python
df_3.resample('M').max()
```

```python
df_3.resample('M').min()
```

```python
# agg関数で集計を1度に行う
df_3.resample('Q').agg(["sum","mean","max","min"])
```

### 曜日ごとの集計

```python
df_3.index.weekday
# 各日付の曜日の取得 6は日曜、0は月曜
```

```python
df_3[df_3.index.weekday == 0]
```

```python
df_3[df_3.index.weekday == 0].mean()
```

```python
# 各曜日事の集計を同時に表示
```

```python
# 各曜日事の集計
df_4 = df_3.set_index(df_3.index.weekday)
df_4.index.name = "曜日番号"
df_4.sum(level="曜日番号").sort_index()
```
