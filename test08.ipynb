#### データ部分の並び替え

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 10)
```

```python
df = pd.read_excel('sample.xlsx', sheet_name='実績管理表')
df # 複数あるシート内の内、「実績管理表」を抽出
```

```python
df.sort_values(by='売上金額') # 昇順
```

```python
df.sort_values(by='売上金額', ascending=False) # 降順
```

```python
df.sort_values(by='売上金額', ascending=True) # デフォルトでascendingはTrue
```

```python
df.sort_values(by=['氏名','売上金額']) # 氏名と売上を昇順に並び替え
```

```python
df.sort_values(by=['氏名','売上金額'], ascending=[True,False])
```

```python
df.sort_values(by=['性別','氏名','売上金額'], ascending=[False,True,False])
```

```python
df = pd.read_excel('sample.xlsx',sheet_name='実績管理表_空白')
df
```

```python
df = pd.read_excel('sample.xlsx',sheet_name='実績管理表_空白')
df.sort_values(by='氏名', ascending=True) # 昇順は欠損値1番下
```

```python
df.sort_values(by='氏名', ascending=False) # 降順でも欠損値1番下
```

```python
df.sort_values(by='氏名', ascending=False, na_position='first')
```

```python
df_ax = pd.DataFrame({'col01':[1,2,3],'col02':[4,5,6],'col03':[7,8,9]}
                    ,index=['idx01','idx02','idx03'])
df_ax # indexでも並び替え可能 しかし数字と文字が混在している場合は不可
```

```python
df_ax.sort_values(by='idx01', ascending=False, axis=1)
# idx01 を降順に並び替え
```

```python
df_ax.sort_values(by='col01', ascending=False, axis=0)
#col01 を降順に並び替え
```

```python
df_ax.sort_values(by='idx02', ascending=False, axis=1)
```

```python
df_ax
```

```python
df_ax.sort_values(by='col01', ascending=False)
```

```python
df_ax # 元のデーターフレームのままで並び替え出来ていない
```

```python
df_ax.sort_values(by='col01', ascending=False, inplace=True)
```

```python
df_ax
```

```python
df_ax = df_ax.sort_values(by='col01', ascending=True)
#inplace を使用しない方法
```

```python
df_ax
```

```python
# sort_index indexを並び替えする為by='columns'必要なし
```

```python
df_ax.sort_index(ascending=False)
```

```python
df_ax.sort_index(ascending=False,inplace=True,na_position='first')
```

```python
df_ax # 元々昇順だったものをinplaceで降順に変更
```

```python
df_ax.columns
```

```python
df_ax.columns.sort_values(ascending=False)
```

```python
df_ax.sort_index(axis=1,ascending=False) # カラム名を降順
```
